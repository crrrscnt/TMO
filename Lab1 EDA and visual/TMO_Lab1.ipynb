```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

# Датасет с ирисами
sepal length - длина чашелистника

sepal width - ширина чашелистника

petal length - длина лепестка

petal width - ширина лепестка

target - целевой атрибут (принадлежность классу)


       0 -  Iris-Setosa

       1 - Iris-Versicolour

       2 - Iris-Virginica



```python
iris_sk = load_iris()
```

```python
df = pd.DataFrame(data = iris_sk.data, columns = iris_sk.feature_names)
```

```python
df["target"] = iris_sk.target
df
```

# Визуализация

```python
sns.pairplot(df, hue="target")
```

```python
# Распределение параметра Petal Length, width и sepal l, w сгруппированные по Target.
sns.violinplot(x='target', y= 'petal length (cm)', data=df, label="Длина лепестка")
sns.violinplot(x='target', y= 'petal width (cm)', data=df, label="Ширина лепестка")
sns.violinplot(x='target', y= 'sepal length (cm)', data=df, label="Длина чашелистника")
sns.violinplot(x='target', y= 'sepal width (cm)', data=df, label="Ширина чашелистника")
```
